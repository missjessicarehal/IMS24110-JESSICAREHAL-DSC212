```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
import pandas as pd

# -----------------------------
# 1. SETUP
# -----------------------------
G = nx.karate_club_graph()
n = G.number_of_nodes()
m = G.number_of_edges()
global_2m = 2 * m

initial_nodes = sorted(G.nodes())
community_tracker = {node: 0 for node in initial_nodes}
next_community_id = 1
iteration_count = 1
all_metrics_by_iteration = {}
modularity_over_time = []

# Fixed layout for consistency
pos = nx.spring_layout(G, seed=42)

# -----------------------------
# 2. HELPER FUNCTIONS
# -----------------------------
def draw_graph(G, pos, community_tracker, iteration, title):
    """Visualize the graph with colors by community."""
    unique_communities = sorted(set(community_tracker.values()))
    num_colors = max(1, len(unique_communities))
    cmap = plt.cm.get_cmap('Spectral', num_colors)
    node_colors = [cmap(community_tracker[node] / num_colors) for node in G.nodes()]

    plt.figure(figsize=(8, 6))
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=500)
    nx.draw_networkx_edges(G, pos, alpha=0.5)
    nx.draw_networkx_labels(G, pos, labels={n: n for n in G.nodes()}, font_color='black')
    plt.title(f"{title}\nIteration {iteration}")
    plt.axis('off')
    plt.show()


def compute_modularity_matrix(subgraph_nodes):
    """Compute restricted modularity matrix B^(C)."""
    nodes_C = sorted(subgraph_nodes)
    A_C = nx.adjacency_matrix(G, nodelist=nodes_C).todense()
    k_C = np.array([G.degree(node) for node in nodes_C])
    B_C = A_C - np.outer(k_C, k_C) / global_2m
    return B_C, nodes_C

# -----------------------------
# 3. RECURSIVE SPECTRAL SPLIT
# -----------------------------
def recursive_split(current_nodes, community_id):
    """Recursive modularity-based spectral bisection (no stopping condition)."""
    global iteration_count, next_community_id

    if len(current_nodes) <= 1:
        return

    B_C, nodes_C = compute_modularity_matrix(current_nodes)
    w, v = eigh(B_C)
    lambda_1 = w[-1]
    u_1 = v[:, -1]

    if lambda_1 <= 1e-5:
        return

    # Split by sign of eigenvector (fallback to median if needed)
    C_plus = {nodes_C[i] for i, val in enumerate(u_1) if val > 0}
    C_minus = {nodes_C[i] for i, val in enumerate(u_1) if val <= 0}
    if not C_plus or not C_minus:
        median = np.median(u_1)
        C_plus = {nodes_C[i] for i, val in enumerate(u_1) if val > median}
        C_minus = {nodes_C[i] for i, val in enumerate(u_1) if val <= median}
    if not C_plus or not C_minus:
        return

    # Assign new community ID
    new_comm_id = next_community_id
    next_community_id += 1
    for node in C_plus:
        community_tracker[node] = new_comm_id

    # Compute modularity
    partition = {}
    for n, cid in community_tracker.items():
        partition.setdefault(cid, []).append(n)
    Q = nx.community.modularity(G, partition.values())
    modularity_over_time.append((iteration_count, Q))

    # Compute metrics on full graph
    metrics = {
        'degree_centrality': nx.degree_centrality(G),
        'betweenness_centrality': nx.betweenness_centrality(G),
        'closeness_centrality': nx.closeness_centrality(G),
        'clustering': nx.clustering(G)
    }
    all_metrics_by_iteration[iteration_count] = metrics


    draw_graph(G, pos, community_tracker, iteration_count, f"Split {iteration_count}")
    print(f" Split {iteration_count}: Community {community_id} â†’ "
          f"{community_id} ({len(C_minus)}) & {new_comm_id} ({len(C_plus)}) | Q = {Q:.4f}")

    iteration_count += 1


    recursive_split(C_plus, new_comm_id)
    recursive_split(C_minus, community_id)


print("Starting Recursive Spectral Modularity Partitioning...")
draw_graph(G, pos, community_tracker, 0, "Initial Graph (Single Community)")


initial_Q = nx.community.modularity(G, [set(G.nodes())])
modularity_over_time.append((0, initial_Q))
all_metrics_by_iteration[0] = {
    'degree_centrality': nx.degree_centrality(G),
    'betweenness_centrality': nx.betweenness_centrality(G),
    'closeness_centrality': nx.closeness_centrality(G),
    'clustering': nx.clustering(G)
}

recursive_split(set(G.nodes()), 0)


print("\n--- Final Communities ---")
final_partition = {}
for node, cid in community_tracker.items():
    final_partition.setdefault(cid, []).append(node)
for cid, nodes in final_partition.items():
    print(f"Community {cid} (Size {len(nodes)}): {sorted(nodes)}")

draw_graph(G, pos, community_tracker, iteration_count - 1, "Final Community Structure")


print("\n--- Generating Metric Evolution Plots ---")
data_records = []
for iteration, metrics_data in all_metrics_by_iteration.items():
    for metric_name, node_values in metrics_data.items():
        for node, value in node_values.items():
            data_records.append({
                'iteration': iteration,
                'node': node,
                'metric_name': metric_name,
                'value': value
            })
metrics_df = pd.DataFrame(data_records)

metrics_to_plot = ['degree_centrality', 'betweenness_centrality', 'closeness_centrality', 'clustering']

plt.figure(figsize=(15, 12))
for i, metric in enumerate(metrics_to_plot):
    plt.subplot(2, 2, i + 1)
    metric_data = metrics_df[metrics_df['metric_name'] == metric]

    for node in sorted(G.nodes()):
        node_data = metric_data[metric_data['node'] == node].sort_values(by='iteration')
        if node_data.empty:
            continue
        if node == 0:
            plt.plot(node_data['iteration'], node_data['value'], 'r-', linewidth=2.5, label="Node 0 (Mr. Hi)")
        elif node == 33:
            plt.plot(node_data['iteration'], node_data['value'], 'b--', linewidth=2.5, label="Node 33 (President)")
        else:
            plt.plot(node_data['iteration'], node_data['value'], 'k-', linewidth=0.6, alpha=0.6)

    plt.title(f"Evolution of {metric.replace('_', ' ').title()}")
    plt.xlabel("Iteration")
    plt.ylabel(metric.replace('_', ' ').title())
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(loc='best', fontsize='small')

plt.tight_layout()
plt.show()

# -----------------------------
# 6. MODULARITY EVOLUTION
# -----------------------------
modularity_df = pd.DataFrame(modularity_over_time, columns=["iteration", "modularity"])
max_idx = modularity_df["modularity"].idxmax()
best_iter = modularity_df.loc[max_idx, "iteration"]
best_Q = modularity_df.loc[max_idx, "modularity"]

plt.figure(figsize=(8, 5))
plt.plot(modularity_df["iteration"], modularity_df["modularity"], "go-", linewidth=2.5)
plt.axvline(best_iter, color='r', linestyle='--', label=f"Peak Q={best_Q:.3f} at iteration {best_iter}")
plt.title("Modularity Across Iterations")
plt.xlabel("Iteration")
plt.ylabel("Modularity (Q)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# -----------------------------
# 7. METRIC MATRIX SUMMARY
# -----------------------------
summary = []
for iteration, data in all_metrics_by_iteration.items():
    row = {'iteration': iteration}
    for metric_name, node_values in data.items():
        row[f"{metric_name}_avg"] = np.mean(list(node_values.values()))
    summary.append(row)
summary_df = pd.DataFrame(summary).sort_values(by='iteration')
print("\n--- Average Metric Values per Iteration ---")
display(summary_df)

```
